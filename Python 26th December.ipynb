```python
def area_of_rectangle():
    length=90
    breadth=40
    area=length*breadth
    print("area of rectangle=",area)
    print("area of rectangle with length",length,"and breadth",breadth,"is",area)
```

```python
def circumference_of_circle():
    radius=25
    pi=3.14
    circum=2*pi*radius
    print("circumference of circle is",circum)
    print("circumference of circle with radius=",radius,"is",circum )
    
```

```python
circumference_of_circle()
```

```python
#area of circle------- pi radius**2
def areaofcircle():
    pi=3.14
    radius=int(input("Enter radius value"))
    area=pi*(radius**2)
    print("Area=",area)
```
